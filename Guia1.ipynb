```python
import numpy as np
import plotly.graph_objects as go

n = 40

# Generar 40 datos distribuidos uniformemente entre 0 y 1
datos_uniformes = np.random.uniform(0, 1, n)

# Crear el histograma
fig = go.Figure(data=[go.Histogram(x=datos_uniformes, 
                                   marker=dict(color='salmon', 
                                               line=dict(color='black', width=0.5)))]) 
fig.update_layout(
    title='Histograma de Distribución Uniforme',
    xaxis_title='Valor',
    yaxis_title='Frecuencia',
    height = 500,
    width = 600
)
fig.show()
```

```python
from scipy.stats import rayleigh

sigma = 2

# Generamos datos de distribución Rayleigh
datos_rayleigh = np.random.rayleigh(scale=sigma, size=n)

# Calculamos la función de densidad de probabilidad
x_rayleigh = np.linspace(0, max(datos_rayleigh)+sigma, 1000)
pdf_rayleigh = rayleigh.pdf(x_rayleigh, scale=sigma)

# Creamos las figuras para las gráficas
fig_rayleigh = go.Figure()

# Añadir histograma y Densidad
fig_rayleigh.add_histogram(x=datos_rayleigh, 
                             histnorm='probability density', 
                             name='Histograma', 
                             opacity=0.6, 
                             marker=dict(color='salmon', 
                                         line=dict(color='black', width=0.5)))
fig_rayleigh.add_scatter(x=x_rayleigh, y=pdf_rayleigh, mode='lines', 
                           name='Funcion de densidad', line=dict(color='black'))

# Ajustes de diseño
fig_rayleigh.update_layout(
    title='Distribución Rayleigh (σ=2)',
    xaxis_title='Valor',
    yaxis_title='Densidad',
    width=600,
    height=500
)

# Mostramos la gráfica

fig_rayleigh.show()
```

```python
from scipy.stats import weibull_min

beta = 1/3
sigma = 2

# Generamos datos de distribución Rayleigh
datos_weibull = np.random.weibull(a=sigma, size=n) * beta

# Calculamos la función de densidad de probabilidad
x_weibull = np.linspace(0, max(datos_weibull)+beta, 1000)
pdf_weibull = weibull_min.pdf(x_weibull, sigma, scale=beta)

# Creamos las figuras para las gráficas
fig_weibull = go.Figure()

# Añadir histograma y Densidad
fig_weibull.add_histogram(x=datos_weibull, 
                             histnorm='probability density', 
                             name='Histograma', 
                             opacity=0.6, 
                             marker=dict(color='salmon', 
                                         line=dict(color='black', width=0.5)))
fig_weibull.add_scatter(x=x_weibull, y=pdf_weibull, mode='lines', 
                           name='Funcion de densidad', line=dict(color='black'))

# Ajustes de diseño
fig_weibull.update_layout(
    title='Distribución Weibull (α=2, β=1/3)',
    xaxis_title='Valor',
    yaxis_title='Densidad',
    width=600,
    height=500
)

# Mostramos la gráfica
fig_weibull.show()
```
